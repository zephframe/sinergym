```python
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path
```

```python
rewards = []
qtb = []
for path in Path("./training_results/").rglob("*.npz"):
    print("Reading ", path)
    with np.load(path) as data:
        print(data.files)
        simplified_path = str(path).replace("training_results/", "")
        for item in data.files:
            print(item)
            # print(data[item])
            # if (item == "qtb"):
            #     qtb = data[item]
            if (item == "rewards"):
                # rewards = np.concatenate((rewards, data[item]))
                rewards.append({"title": simplified_path, "data": data[item]})
```
