### Q-Learning with HNP Agent

```python
import numpy as np

from hnp.agents import QLearningAgent, FixedActionAgent
from hnp.environment import ObservationWrapper, create_env
```

```python
config = {
    "agent": {
        "num_episodes": 550,
        "horizon": 24,
        "gamma": 0.99,
        "num_tiles": 20,
        "initial_epsilon": 1,
        "epsilon_annealing": 0.999,
        "learning_rate": 0.1,
        "learning_rate_annealing": 0.999
    },
    "env": {
        "name": "Eplus-office-mixed-discrete-stochastic-v1",
        "normalize": True,
        "obs_to_keep": [],
        "mask": [0, 0, 0, 0]
    }
}

obs_to_keep = np.array(config["env"]["obs_to_keep"])
mask = np.array(config["env"]["mask"])
```

```python
env = create_env(config["env"])
env = ObservationWrapper(env, obs_to_keep)

agent = QLearningAgent(
    env,
    config["agent"],
    mask,
)
```

```python
agent.train()
agent.save_results()
env.close()
```
